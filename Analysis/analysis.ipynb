# Did National Happiness Levels Impact COVID-19 Mortality Rates?

```python
# Import Modules
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# File Handling

```python
# Import final csv file for analysis as df
file_path = os.path.join("..", "Resources", "AnalysisData.csv")

# Read analysis csv and store as pandas df
df = pd.read_csv(file_path)

# Preview panel df
df.head()
```

# Exploring Data Relationships

## Correlation Table

```python
# Run correlations between variables
main_corr_df = df.corr()

main_corr_df = main_corr_df.iloc[[1, 2, 3, 4, 5, 6, 7, 10, 11, 12, 13, 14], [1, 10, 14]]

main_corr_df.sort_values('COVID-19 deaths per 100,000 population in 2020')
```

## Descriptive Statistics

```python
# Show distribution of COVID-19 deaths per 100k across nations
df.hist(column="COVID-19 deaths per 100,000 population in 2020", bins=20)
plt.title('Distribution of Country-Level 2020 COVID-19 Deaths per 100k')
plt.show()
```

```python
# Show distribution of COVID-19 deaths per 100k across nations
df.hist(column="2019 Life Ladder", bins=10)
plt.title('Distribution of Country-Level 2019 Life Ladder (Happiness) Metric')
plt.xlim(0, 10)
plt.show()
```

## Correlation Scatterplots

### Exploring the Meaning of Happiness

```python
# Show how happiness relates to other variables (create scatterplots against same variables below)
# Create scatter plot comparing happiness to age
df.plot(kind='scatter',x='2019 Life Ladder',y='Median age',color='black')
plt.title('2019 Life Ladder Metric (Happiness) vs. Median Age')
plt.xlim(0, 10)
plt.ylim(0, 55)

# Display the chart
plt.show()
```

```python
# Create scatter plot comparing happiness to life expectancy
df.plot(kind='scatter',x='2019 Life Ladder',y='2019 Healthy life expectancy at birth',color='black')
plt.title('2019 Life Ladder Metric (Happiness) vs. 2019 Life Expectancy at Birth')
plt.xlim(0, 10)
plt.ylim(0, 80)

# Display the chart
plt.show()
```

```python
# Create scatter plot comparing happiness to GDP
df.plot(kind='scatter',x='2019 Life Ladder',y='2019 Log GDP per capita',color='black')
plt.title('2019 Life Ladder Metric (Happiness) vs. 2019 Log of GDP per Capita')
plt.xlim(0, 10)
plt.ylim(0, 13)

# Display the chart
plt.show()
```

```python
# Create scatter plot comparing happiness to institutional trust
df.plot(kind='scatter',x='2019 Life Ladder',y='Index of institutional trust',color='black')
plt.title('2019 Life Ladder Metric (Happiness) vs. Index of Institutional Trust')
plt.xlim(0, 10)
plt.ylim(0, 1)

# Display the chart
plt.show()
```

### Examining Potential Correlations Between COVID-19 Deaths and Happiness (and Other Quality of Life Measures)
